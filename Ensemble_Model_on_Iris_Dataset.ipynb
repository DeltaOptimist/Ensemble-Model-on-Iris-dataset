```python
# Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

```

```python
# Load dataset
data = load_iris()
X, y = data.data, data.target

```

```python
print(X.shape)
print(y.shape)
```

```python
# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# Define individual models
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
gb_clf = GradientBoostingClassifier(n_estimators=100, random_state=42)
svc_clf = SVC(probability=True, random_state=42)
```

```python
# Combine models into a voting classifier
voting_clf = VotingClassifier(
    estimators=[('rf', rf_clf), ('gb', gb_clf), ('svc', svc_clf)],
    voting='soft'  # Soft voting uses predicted probabilities for the classes
)
```

```python
# Train the ensemble model
voting_clf.fit(X_train, y_train)
```

```python
# Make predictions and evaluate the model
y_pred = voting_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Ensemble Model Accuracy:", accuracy)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
```

```python
# Define a pipeline for scaling and dimensionality reduction
pipeline = Pipeline([
    ('scaler', StandardScaler()),        # Feature Scaling
    ('pca', PCA(n_components=2)),        # Dimensionality Reduction
])

```

```python
# Transform the features
X_train_transformed = pipeline.fit_transform(X_train)
X_test_transformed = pipeline.transform(X_test)

```

```python
# Fit the voting classifier with transformed data
voting_clf.fit(X_train_transformed, y_train)
y_pred = voting_clf.predict(X_test_transformed)
accuracy = accuracy_score(y_test, y_pred)

print("Transformed Ensemble Model Accuracy:", accuracy)
```
